```python
import praw
import os
import requests
import shutil
from PIL import Image
from io import BytesIO
```

```python
user_agent = 'Scraper 1.0 ShadiFarzankia'
reddit = praw.Reddit(
    client_id = 'JtxyXaucGyV3CFyTDnE3tg',
    client_secret = 'cUAUSaeRJwkdQE4xhfQYB0GeMzyvOg',
    user_agent=user_agent
)
```

```python
# Create directory to save images
os.makedirs('images', exist_ok=True)
```

```python
# List of fashion subreddits
subreddits = ["streetwear", "fashion", "femalefashionadvice"]

# Loop through each subreddit
for sub_name in subreddits:
    subreddit = reddit.subreddit(sub_name)
    print(f"Scraping r/{sub_name}...")

    counter = 1  # start numbering from 1

    for post in subreddit.top(limit=5000):
        if post.url.lower().endswith(('jpg', 'jpeg', 'png', 'webp')):
            image_url = post.url
            jpg_name = f"{sub_name}_{counter}.jpg"
            jpg_path = os.path.join("images", jpg_name)

            try:
                response = requests.get(image_url, timeout=10)
                response.raise_for_status()
                image = Image.open(BytesIO(response.content)).convert("RGB")
                image.save(jpg_path, "JPEG")
                print(f"✅ Saved: {jpg_name}")
                counter += 1  # increment only on successful save
            except Exception as e:
                print(f"❌ Failed to process {image_url}: {e}")
```

```python
# Zip the folder
shutil.make_archive('images_backup', 'zip', 'images')
print("Zipped folder saved as images_backup.zip")
```
